## Car Analysis using Pandas Dataframe

```python
import pandas as pd
```

```python
car = pd.read_csv("Car_dataset.csv")
```

```python
car.head()
```

```python
car.shape
```

### Data Cleaning: filling in the null values with the mean of that column.

```python
car.isnull().sum()
```

```python
# Fill NaN values only for numeric columns with their respective mean
car.fillna(car.select_dtypes(include=['number']).mean(), inplace=True)
```

### Check what are the different types of make aret there in our dataset, and what is the count occurence of each make in the data?

```python
car.head(2)
```

```python
car['Make'].value_counts()
```

### Show all the records where the origin is Asia or Europe

```python
car[car['Origin'].isin(['Asia','Europe'])]
```

### Remove all the records (rows) where the weight is above 4000

```python
car.head(2)
```

```python
car[~(car['Weight'] > 4000)]
```

### Increase all the values of 'MPG_City' column by 3.

```python
car.head(2)
```

```python
car['MPG_City'] = car['MPG_City'].apply(lambda x:x+3)
```
